## Probando el modelo que hicimos :D
Primero vamos a importar las librerías que vamos a ocupar para esto...

```python
from keras.preprocessing import image 
from keras.utils import load_img
from keras.utils import img_to_array
from keras import models
import numpy as np
import matplotlib.pyplot as plt
```

### Seleccionando el modelo
Ya que generé dos modelos distintos, podemos comentar y descomentar lo siguiente, el mejor fue el modelo normal, ya que tuve problemas para implementar el modelo usando aumento de datos. Pasaron cosas muy extrañas.

```python
model = models.load_model("./models/tom_and_jerry_1.h5") # Modelo normal
# model = models.load_model("./models/tom_and_jerry_1.h5") # Modelo con aumento de datos
```

### Cargando las imágenes
A continuación podemos descomentar la imagen con la que queremos probar el modelo.   

Significado de los resultados:   
1 = TOM   
0 = JERRY

```python
# Descomenta la imagen que quieres usar:

test_img = './images/j.jpg'
# test_img = './images/j2.jpg'
# test_img = './images/j3.jpg'
# test_img = './images/j4.jpg'
# test_img = './images/t.jpg'
# test_img = './images/t2.jpg'
# test_img = './images/t3.jpg'
# test_img = './images/spike.jpg'
# test_img = './images/tom_y_black.jpeg'
# test_img = './images/tom_y_jerry.jpg'


img_path = test_img 

# Preprocesado
img = load_img(img_path,  target_size=(256,256))
img_tensor = img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis = 0)
img_tensor /= 255.

# Respuesta del modelo
confidence = model.predict(img_tensor)
predict_class = (confidence > 0.5).astype("int32")
print (confidence)
print ("class ", predict_class[0][0], "confindence", )

# Predicción 
if predict_class[0][0] == 0:
  print("JERRY")
else:
  print("TOM")

# Plot de la imagen a predecir
plt.imshow(img_tensor[0])
plt.show()
```
